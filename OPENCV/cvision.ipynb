```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = np.full((8, 16,3),100, dtype='uint8')
image[1:3,2:10]=[244,96,48]
image[3:5,2:10]=[255,255,255]
image[5:7,2:10]=[0,100,0]
plt.imshow(image, cmap='jet')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
image=np.full ((8,16,3), 100, dtype='uint8') 
cv2.rectangle(image, (4, 4), (7, 4), (48, 96, 244), -1)  # Red bar (BGR: 0, 0, 255)
cv2.rectangle(image, (4, 5), (7, 5), (255, 255, 255), -1)  # Green bar (BGR: 0, 255, 0)
cv2.rectangle(image, (4, 6), (7, 6), (0, 100, 0), -1)  # Blue bar (BGR: 255, 0, 0)
# cv2.imshow('image', image)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# 4. Display the image
plt.imshow(image_rgb)
plt.show()
cv2.imwrite('image.jpg', image)
cv2.waitKey(100000)
cv2.destroyAllWindows()
```
