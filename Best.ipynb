HOUSE RENT PREDICTOR



```python
import pandas as pd
data=pd.read_csv("House_Rent_Dataset.csv")
data.head()
```

```python
 y =data['Rent']
 x=data.drop(['Rent','Posted On','Floor','Area Locality'],axis=1)
```

```python
import category_encoders as ce
encoder=ce.LeaveOneOutEncoder()
X= encoder.fit_transform(x,y)
```

```python
print(X)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(X)
```

```python
print(x)
```

```python
import pandas as pd 
x=pd.DataFrame(x,columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
new_house = pd.DataFrame({

'BHK': [2],

'Size': [1200],

'Area Type': ['Super Area'],

'City': ['Delhi'],

'Furnishing Status': ['Semi-Furnished'],

'Tenant Preferred': ['Family'],

'Bathroom': [2],

'Point of Contact': ['Contact Owner']

})

new_encoded = encoder.transform(new_house) 
new_scaled = scaler.transform(new_encoded) 
prediction = model.predict(new_scaled) 
print("Predicted Rent:", prediction[0])
```

```python
import joblib

# Assuming you have already trained these three objects in your code:
# 1. your_encoder  (e.g., OneHotEncoder or LabelEncoder)
# 2. your_scaler   (e.g., StandardScaler or MinMaxScaler)
# 3. your_model    (e.g., LinearRegression, RandomForestRegressor, etc.)

# Create a dictionary matching what hpp.py expects to load
saved_data = {
    "encoder": encoder,
    "scaler": scaler,
    "model": model
}

# Save the dictionary to a .pkl file
joblib.dump(saved_data, "house_rent_prediction.pkl")

print("Model saved successfully!")
```

```python
import joblib
import pandas as pd
hpp=joblib.load("house_rent_prediction.pkl")
encoder=hpp["encoder"]
scaler=hpp["scaler"]
model=hpp["model"]
new_house = pd.DataFrame({
'BHK': [2],
'Size': [1200],
'Area Type': ['Super Area'],
'City': ['Delhi'],
'Furnishing Status': ['Semi-Furnished'],
'Tenant Preferred': ['Family'],
'Bathroom': [2],
'Point of Contact': ['Contact Owner']
})

new_encoded = encoder.transform(new_house) 
new_scaled = scaler.transform(new_encoded) 
prediction = model.predict(new_scaled) 
print("Predicted Rent:", prediction[0])
```

```python
from sklearn.pipeline import Pipeline 
pipeline = Pipeline([
    ("encoder", ce.LeaveOneOutEncoder()),
    ("scaler", MinMaxScaler()), 
    ("model", LinearRegression()) ]) 
pipeline.fit(x_train, y_train) 
```

```python
from sklearn.metrics import r2_score, mean_absolute_error 
y_pred = pipeline.predict(x_test) 
print("R² Score :", r2_score(y_test, y_pred)) 
print("MAE :", mean_absolute_error(y_test, y_pred)) 
```
